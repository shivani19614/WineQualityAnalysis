```python
import pandas as pd

# Load the red wine dataset
red_wine_data = pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-red.csv", sep=';')

# Load the white wine dataset
white_wine_data = pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-white.csv", sep=';')

# Display the first few rows of each dataset
print("Red Wine Data:")
print(red_wine_data.head())

print("\nWhite Wine Data:")
print(white_wine_data.head())
```

```python
import pandas as pd

# Load the white wine dataset
white_wine_url = "https://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-white.csv"
white_wine_data = pd.read_csv(white_wine_url, delimiter=";")

# Load the red wine dataset
red_wine_url = "https://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-red.csv"
red_wine_data = pd.read_csv(red_wine_url, delimiter=";")

# Add a 'type' column to distinguish between white and red wines
white_wine_data['type'] = 'white'
red_wine_data['type'] = 'red'

# Concatenate both datasets
wine_data = pd.concat([white_wine_data, red_wine_data], ignore_index=True)


```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Combine red and white wine datasets for additional visualizations
combined_data = pd.concat([red_wine_data, white_wine_data], axis=0, ignore_index=True)

# Visualize the distribution of alcohol content for red and white wines
plt.figure(figsize=(10, 6))
sns.histplot(data=combined_data, x='alcohol', hue='type', kde=True, palette=['crimson', 'lightseagreen'])
plt.title("Distribution of Alcohol Content by Wine Type")
plt.xlabel("Alcohol Content")
plt.ylabel("Count")
plt.legend(title="Wine Type")
plt.show()

# Visualize the distribution of acidity levels (fixed acidity) for red and white wines
plt.figure(figsize=(10, 6))
sns.histplot(data=combined_data, x='fixed acidity', hue='type', kde=True, palette=['crimson', 'lightseagreen'])
plt.title("Distribution of Fixed Acidity by Wine Type")
plt.xlabel("Fixed Acidity")
plt.ylabel("Count")
plt.legend(title="Wine Type")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create a scatter plot for red wine
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
sns.scatterplot(data=red_wine_data, x='alcohol', y='volatile acidity', hue='quality', palette='coolwarm')
plt.title("Red Wine: Quality vs Alcohol vs Volatile Acidity")

# Create a scatter plot for white wine
plt.subplot(1, 2, 2)
sns.scatterplot(data=white_wine_data, x='alcohol', y='volatile acidity', hue='quality', palette='viridis')
plt.title("White Wine: Quality vs Alcohol vs Volatile Acidity")

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the figure size
plt.figure(figsize=(12, 6))

# Plot the distribution of white wine quality scores
plt.subplot(1, 2, 1)
sns.countplot(data=wine_data[wine_data['type'] == 'white'], x='quality', palette='Set3')
plt.title("Distribution of White Wine Quality Scores")
plt.xlabel("Quality")
plt.ylabel("Count")

# Plot the distribution of red wine quality scores
plt.subplot(1, 2, 2)
sns.countplot(data=wine_data[wine_data['type'] == 'red'], x='quality', palette='Set3')
plt.title("Distribution of Red Wine Quality Scores")
plt.xlabel("Quality")
plt.ylabel("Count")

# Adjust the spacing between subplots
plt.tight_layout()

# Show the plots
plt.show()
```

```python
# Calculate percentage distribution of wine quality scores for red wine
red_quality_distribution = red_wine_data['quality'].value_counts(normalize=True).sort_index()

# Calculate percentage distribution of wine quality scores for white wine
white_quality_distribution = white_wine_data['quality'].value_counts(normalize=True).sort_index()

print("Red Wine Quality Distribution:")
print(red_quality_distribution)
print("\nWhite Wine Quality Distribution:")
print(white_quality_distribution)

```

```python
# Calculate correlation matrix for red wine
red_correlation_matrix = red_wine_data.corr()

# Visualize the correlation matrix as a heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(data=red_correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title("Correlation Heatmap (Red Wine)")
plt.show()
```

```python
from sklearn.ensemble import RandomForestRegressor

# Split the data into features (X) and target (y)
X_red = red_wine_data.drop('quality', axis=1)
y_red = red_wine_data['quality']

# Fit a Random Forest model to determine feature importance
rf_model = RandomForestRegressor(n_estimators=100, random_state=42)
rf_model.fit(X_red, y_red)

# Get feature importances
feature_importances = pd.Series(rf_model.feature_importances_, index=X_red.columns)

# Sort and visualize feature importances
sorted_feature_importances = feature_importances.sort_values(ascending=False)
plt.figure(figsize=(10, 6))
sns.barplot(x=sorted_feature_importances.values, y=sorted_feature_importances.index, palette='muted')
plt.title("Feature Importance (Red Wine)")
plt.xlabel("Importance")
plt.ylabel("Features")
plt.show()
```
